```python
import re
import json
```

```python
with open('words.txt', 'r', encoding='utf8') as file:
    dictionary = file.read()

print(dictionary)
```

```python
dictionary = re.split(r'\n', dictionary)
```

```python
single_gram = []
bi_gram = []
tri_gram = []
quad_gram = []

for word in dictionary:
    sub_word = word.split()
    if len(sub_word) == 1:
        single_gram.append(word)
    elif len(sub_word) == 2:
        bi_gram.append(word)
    elif len(sub_word) == 3:
        tri_gram.append(word)
    elif len(sub_word) == 4:
        quad_gram.append(word)
```

```python
corpus = {'single': single_gram,
          'bi': bi_gram,
          'tri': tri_gram,
          'quad': quad_gram}
with open('corpus.json', 'w', encoding='utf8') as js:
    json.dump(corpus, js, ensure_ascii=False)
```
